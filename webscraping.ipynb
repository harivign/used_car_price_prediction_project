```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Ensure the browser window remains open
driver = webdriver.Chrome()
driver.get('https://www.cardekho.com/used-cars+in+bangalore')
time.sleep(5)

# Scroll until the end of the page
last_height = driver.execute_script("return document.body.scrollHeight")  # Get initial page height

while True:
    # Scroll down
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)  # Wait for the page to load more content

    # Get the new page height
    new_height = driver.execute_script("return document.body.scrollHeight")

    # Break the loop if the page height hasn't changed (end of the page)
    if new_height == last_height:
        time.sleep(5)
        break

    last_height = new_height
    
# Find the elements containing the car details
bengalure = driver.find_elements(By.XPATH, "//div[@class='gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-4 cardColumn']")

# Extract the text content of the elements
bengalure_car_details = [i.text for i in bengalure]

# Close the browser
driver.quit()

# Initialize the list to store structured data
structured_data = []

# Loop through the extracted car details
for detail in bengalure_car_details:
    # Split the details by lines
    lines = detail.split("\n")
    
    # Extract key information
    try:
        # Split the first line into year and car name
        year_and_name = lines[0].split(" ", 1)
        year = year_and_name[0]  # Year
        car_name = year_and_name[1]  # Car name
        
        # Extract mileage, fuel, transmission (safely handle cases where data is missing or incomplete)
        if "•" in lines[1]:
            details = [item.strip() for item in lines[1].split("•")]
            if len(details) == 3:
                mileage, fuel_type, transmission = details
            elif len(details) == 2:
                mileage, fuel_type = details
                transmission = ""
            else:
                mileage = details[0]
                fuel_type = transmission = ""
        else:
            mileage = fuel_type = transmission = ""
        
        mileage = mileage.replace("kms", "").strip()  # Clean mileage text
        
        # Extract price (safely handle missing data)
        price = lines[2].replace("₹", "").replace("Lakh", "").strip() if len(lines) > 2 else ""
        
        # Extract location (last line)
        location = lines[-1] if len(lines) > 3 else ""
        
        # Append structured data
        structured_data.append({
            "Year": year,
            "Car Name": car_name,
            "Mileage (kms)": mileage,
            "Fuel Type": fuel_type,
            "Transmission": transmission,
            "Price (Lakh)": price,
            "Location": location
        })
    except Exception as e:
        print(f"Error processing detail: {detail}\nError: {e}")

# Create DataFrame
car_df = pd.DataFrame(structured_data)

# Save to CSV
car_df.to_csv('bengalure_car_details.csv', index=False)

# Print DataFrame
print(car_df)
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://www.cardekho.com/used-cars+in+delhi-ncr") 
driver.maximize_window()
time.sleep(5)  # Allow page to load

# Scroll until the end of the page
last_height = driver.execute_script("return document.body.scrollHeight")  # Get initial page height
scroll_attempts = 0

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)  # Allow time for new content to load
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    if new_height == last_height:
        scroll_attempts += 1
        if scroll_attempts > 3:  # Stop after 3 failed attempts to load more content
            break
    else:
        scroll_attempts = 0  # Reset if new content loads

    last_height = new_height

try:
    # Find elements containing car details
    car_elements = driver.find_elements(By.XPATH, "//div[contains(@class,'cardColumn')]")

    # Initialize list for structured data
    structured_data = []

    for car in car_elements:
        details = car.text.split("\n")  # Extract text from each element
        
        try:
            # Extract Year and Car Name
            year_and_name = details[0].split(" ", 1)  # Year is first part
            year = year_and_name[0]
            car_name = year_and_name[1] if len(year_and_name) > 1 else ""

            # Extract Mileage, Fuel Type, Transmission
            mileage, fuel_type, transmission = "", "", ""

            if "•" in details[1]:
                specs = [x.strip() for x in details[1].split("•")]
                if len(specs) >= 3:
                    mileage, fuel_type, transmission = specs[:3]
                elif len(specs) == 2:
                    mileage, fuel_type = specs
                elif len(specs) == 1:
                    mileage = specs[0]

            # Clean Mileage Data
            mileage = mileage.replace("kms", "").strip()

            # Extract Price (handling missing values)
            price = ""
            for item in details:
                if "₹" in item:
                    price = item.replace("₹", "").replace("Lakh", "").strip()
                    break

            # Extract Location (Assumed last line contains location)
            location = details[-1] if len(details) > 3 else ""

            structured_data.append({
                "Year": year,
                "Car Name": car_name,
                "Mileage (kms)": mileage,
                "Fuel Type": fuel_type,
                "Transmission": transmission,
                "Price (Lakh)": price,
                "Location": location
            })

        except Exception as e:
            print(f"Error processing car details: {details}\nError: {e}")

finally:
    driver.quit()  # Always close the browser

# Create DataFrame
car_df = pd.DataFrame(structured_data)

# Save to CSV
car_df.to_csv('delhi_used_cars.csv', index=False)
print(car_df)
```
